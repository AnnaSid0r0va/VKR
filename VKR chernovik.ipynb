```python
%matplotlib inline
import matplotlib as mpl
import seaborn as sns
```

```python
spec = mpl.gridspec.GridSpec(
    ncols=11, nrows=15, height_ratios=[1, 4] * 10, hspace=0.55
    )
```

```python
fig = plt.figure(figsize=(2, 4))

for i in range(5):
    for j in range(3):
        ax = fig.add_subplot(spec[6*i+j])
        sns.boxplot(data=df, ax=ax,
                   x=df.columns[3*i+j],
                   palette = 'Set2') .set(xlabel=None)
        
        ax = fig.add_subplot(spec[6*i+j+3])
        sns.histplot(x=df.columns[3*i+j],
                    kde=True, ax=ax,
                    data=df, stat='probability',
                    palette='Set2') .set(ylabel=None)
plt.show()
```
